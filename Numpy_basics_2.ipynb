```python
import numpy as np
arr=np.array([[1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2]])
print("original array:\n",arr)
```

```python
print("Every other row:\n",arr[0:3:2])
```

```python
arr=np.array([1,2,3,4,5,6,7])
print("original array:\n",arr)
print("Return every other element in the array:arr[::2]:",arr[::2])
```

```python
arr=np.array([[1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2]])
temp=arr[:2,:3]
print("\n Array with first 2 rows and 3 column:\n",temp)
```

```python
newarr=arr[[0,1,2,3],[3,2,1,0]]
print("\nElements at indices (0,3),(1,2),(2,1),(3,0):\n",newarr)
```

```python
cond=arr>2
temp=arr[cond]
print("Elements greater than 2:\n",temp)
```

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([1,5,6])
arr3=np.hstack((arr1,arr2))
print(arr3)
```

```python
arr4=np.vstack((arr1,arr2))
print(arr4)
```

```python
arr5=np.dstack((arr1,arr2))
print(arr5)
```

```python
arr=np.array([1,2,3,4,5,6])
arr6=np.array_split(arr,3)
print(arr)
print(arr6)
```
